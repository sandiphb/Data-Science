<b>Question 1: Read the data from the CSV file stored on the URL, using NumPy Library.<br>
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'<br>
Take the input of first four columns and then find the 5 closest samples (stored in
CSV) of the input taken, by using Euclidean distance between input sample and all
samples stored in CSV file. In calculating the distance, ignore the last feature (i.e.
class) of the CSV file. <br>Print the class of these closest samples.</b>

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# iris = np.genfromtxt(url, delimiter=',', dtype='float,'*4+"U20")
iris = pd.read_csv(url)
print(iris)
```

```python
iris.head()
```

```python
iris_new = iris.drop(iris.columns[4],axis=1)
```

```python
iris_new.describe()
```

```python
import math
#input df 
df=[5.84,3.05,3.77,1.20]
def euclidean_distance(i,ds):
    sum_sq = np.square(ds.iat[i,0]-df[0])+np.square(ds.iat[i,1]-df[1])+np.square(ds.iat[i,2]-df[2])+np.square(ds.iat[i,3]-df[3])
    distance = math.sqrt(sum_sq)
    return distance

dist=[]
for i in range(len(iris_new)):
    dist.append(euclidean_distance(i,iris_new))

print(dist)
```

```python
#find 5 closest samples
#index of 5 min distance in dist
l=[]
for i in range(5):
    #storing index of min element
    l.append(dist.index(min(dist)))
    #assigning min as max so that we get next min.
    dist[l[i]]=max(dist)

print(l)
    
```

```python
for i in range(5):
    print(iris.iat[l[i],4])
```
